```python
int_to_float = lambda n: struct.unpack('<f', struct.pack('<I', n))[0]
float_to_int = lambda f: struct.unpack('<I', struct.pack('<f', f))[0]
float_to_bin = lambda x: bin(float_to_int(x))


def build_guess_values(value_type='mantissa', mantissa_nbits=10):
    max_mantissa_nbits = 23
    exponent_nbits = 8
    if value_type == 'mantissa':
        s = 0
        e = 127 << max_mantissa_nbits
        m = np.left_shift(np.arange(0, np.power(2, mantissa_nbits)), max_mantissa_nbits-mantissa_nbits)
    elif value_type == 'exponent':
        s = 0
        m = 1 << (max_mantissa_nbits - 1)
        e = np.left_shift(np.arange(0, np.power(2, exponent_nbits)), max_mantissa_nbits)
    elif value_type == 'sign':
        s = np.left_shift(np.asarray([0, 1]), max_mantissa_nbits+exponent_nbits)
        m = 1 << (max_mantissa_nbits - 1)
        e = 1 << max_mantissa_nbits
    else:
        raise ValueError('the value_type is not supported')
    return np.vectorize(int_to_float)(s | m | e)
```

```python
max_mantissa_nbits = 16
two_powers = 1 / np.asarray([np.power(2.0, i) for i in np.arange(0, max_mantissa_nbits)])


def compute_mantisa(n):
    """
    get the floating number from n.
    :param n:
    :return:
    """
    retval = 0.0
    s = "{0:b}".format(n)
    l = len(s)
    for idx in range(l):
        if s[l-idx-1] == '1':
            retval = retval + two_powers[idx]
    return retval
```

```python
for i in [0, 1, 2, 4, 8, 16]:
    print('%03d' % i, '{:032b}'.format(float_to_int(compute_mantisa(i))))
```

```python
import numpy as np
import pandas as pd
from pprint import pprint
from guess_range import *
from batina_guess_number import *
from guess_range import *
```

```python
def compute_corr(secret_hw, guess_range, known_inputs, guess_val_size=1000):
    low, high = guess_range
    guess_val = np.arange(low, high, (high - low) / (guess_val_size - 1e-127))
    hw = pd.DataFrame(columns=guess_val,
                        data=np.vectorize(hamming_weight)(known_inputs.reshape(-1, 1) * guess_val))
    return hw.corrwith(pd.Series(secret_hw), method='pearson')
```

```python
guess_range=(0.0, 1e0)
# the number of guess values in the guess range
guess_val_size=1000
# generate the known inputs which is used to test the guess values
input_range = 0
known_inputs = np.random.uniform(-10 ** input_range, 10 ** input_range, 1000)

secret_number = np.random.uniform(guess_range[0], guess_range[1]) 
secret_number = 0.812
print('secret_number =', secret_number)
```

```python
secret_hw = mul_hamming_weight(known_inputs, secret_number)
correlations = compute_corr(secret_hw, guess_range, known_inputs, guess_val_size)
correlations.sort_values(ascending=False, inplace=True)
```
